```python
import configparser
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import String, Integer, Float, Boolean, Date, Column
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy import func
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Using Object Relational Mapper (ORM)
# Define the DB tables that we will be using in this code
Base = declarative_base()

class Title(Base):
    __tablename__ = 'Titles'
    emp_no = Column(Integer, primary_key=True)
    title = Column(String(500))
    from_date = Column(Date)
    to_date = Column(Date)

class Salary(Base):
    __tablename__ = 'Salaries'
    emp_no = Column(Integer, primary_key=True)
    salary = Column(Float)
    from_date = Column(Date)
    to_date = Column(Date)
```

```python
# Name of DB to connect. Could be passed as parameter
dbname = 'Employees_db'
```

```python
# Read INI file with necessary data about DB server
config = configparser.ConfigParser()
config.read('db_info.ini')

# assign the postgress section to a variable so we can use it easily in the connection code
postgDB = config['postgress']

# create the connection string to postgress
conn_str = 'postgresql://' + postgDB['user'] + ':' + postgDB['passwd'] + '@' + postgDB['host'] + ':' + \
    postgDB['port'] + '/' + dbname
engine = create_engine(conn_str)

# Now create the session
Session = sessionmaker(bind=engine)
session = Session()
```

```python
# Using Object Relational Mapper (ORM)
results = session.query(Title.title, func.avg(Salary.salary)).join(Salary, Title.emp_no==Salary.emp_no).\
    group_by(Title.title).all()

# Read all results into a dataframe for plotting
avgSal_df = pd.DataFrame(results, columns = ["Title","Avg Salary"])

# Close the session to free up memory
session.close()
```

```python
# Create a bar chart of average salary by title
plt.figure(figsize=(10,5))
plt.bar(avgSal_df['Title'], avgSal_df['Avg Salary'], color=['yellow','red','green','blue','cyan','purple','lightgreen'],\
         edgecolor='black', alpha=0.65, align="center")

# Let provide good labels for each bar
x_pos = np.arange(len(avgSal_df['Title']))
plt.xticks(x_pos, avgSal_df['Title'], rotation=45, ha='right')


# Provide Axis titles and chart title
plt.xlabel("Position Title")
plt.ylabel("Average Salary")
plt.title("Average Salary per Position Title")

plt.tight_layout()
#plt.savefig("../Images/avg_salaries.png")
plt.show()
```
